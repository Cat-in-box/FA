## 2 вопрос.

```python
import pandas as pd

# Чтение файла
data = pd.read_csv('./TOBD21_datasets/datasets_for_task_2/sp500hst.txt', header=None, names=['date', 'Tiker', 'open', 'high', 'low', 'closing', 'smth_else'], parse_dates=[0])
data
```

```python
# Проверка типов
data.dtypes
```

```python
# Поиск средних значений OHLC за 2010 год с группировкой по тикерам
result = data[data['date'].dt.year==2010][['open', 'high', 'low', 'closing']].groupby(data['Tiker']).mean()
result
```

```python
# Запись в файл
result.to_csv('task2_result.csv')
```

## 3 вопрос.

```python
import dask.bag as db
import re

# Подсчет реплик внутри блоков
def counter(data):
    return len(re.findall(r'^\s*[—–-]', data)) # В разных файлах используются разные символы в качестве тире, поэтому перечислила несколько вариантов

# Открытие файла
bag = db.read_text(urlpath='./TOBD21_datasets/datasets_for_task_3/all_k/*', encoding='Windows-1251')
# Применение функции counter и суммирование всех результатов, вызов compute() для старта вычислений
bag.map(counter).sum().compute()
```
