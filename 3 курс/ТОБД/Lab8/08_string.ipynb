1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```
"Название"

1. Шаг 1.
2. Шаг 2.
----------
#тэг1 #тэг2
```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
from bs4 import BeautifulSoup
````

````python
steps = BeautifulSoup(open('../data/steps_sample.xml').read(), 'xml')
steps
````

````python
def show_info(recipe_id):
    current_recipe = recipes[recipes['id']==recipe_id]
    result_str = ''

    recipe_name = current_recipe['name'][0]
    result_str += f'"{recipe_name}"\n'
    
    recipe_steps = []
    for recipe in steps.find_all('recipe'):
        if recipe_id == int(recipe.find('id').get_text()):
            for each in recipe.find_all('step'):
                recipe_steps.append(each.get_text())

    for i in range(len(recipe_steps)):
        result_str += f'{i+1}. {recipe_steps[i]}.\n'
    result_str += '----------\n'
    
    recipe_tags = current_recipe['tags'][0].split(';')
    return result_str + '#{}'.format(' #'.join(recipe_tags))
````
